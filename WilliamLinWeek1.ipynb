Section 2

Elements common to all control tasks:


State (St): The current situation or conditions of the environment. 


Action (At): The possible moves or decisions that the agent can make. 


Reward (Rt): The numerical value received by the agent after taking an action. 


Agent: The entity responsible for observing the state and taking actions. 


Environment: The aspects of the task that the agent cannot control.


Deterministic vs Stochastic: 


Deterministic: same outcomes for a given set of inputs. Ex: Chess 


Stochastic: randomized Ex: Monopoly (dice) 


Policy: 
Function that decides what action to take in a particular state. Can be either
stochastic or deterministic .

Markov Decision Process (stochastic): 
Used to model an environment for an agent to learn


Represented as a mathematical expression: (S, A, R, P) 


S - Set of possible states of the task 


A - Set of actions that could be taken in each of the states 


R - Set of rewards for each (s,a) pair 


P - Probabilities of passing from one state to another when taking each 
possible action 


Markov property states that future state depends only on the current state and 
action taken. 


Goal is to find a policy that maximizes the expected cumulative reward over time. Can be either Finite (Episodic) or Infinite (Continuing)

Trajectory: sequence of states, actions, and rewards that an agent experiences while interacting with an environment. 


Episode: initial state to terminal state 


Reward: immediate result of an action in an environment 
Return: Sum of rewards in an episode

Discount Factor: 
determines importance of future rewards vs immediate rewards. 


γ = [0,1]

γ is closer to 1: agent considers long-term consequences (efficiency)

γ is closer to 0: agent focus on short-term gains.

Bellman Equations: 

Expected Value of a State = Probability of taking each action following the policy multiplied by the return obtained (reward upon reaching the state + discounted factor of the state)
